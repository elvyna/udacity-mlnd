```python
import pandas as pd
import numpy as np
import seaborn as sns
from tqdm import tqdm, tqdm_pandas
import matplotlib.pyplot as plt
%matplotlib inline


plt.style.use('bmh')
```

```python
train = pd.read_csv("../dataset/train_feats_init.csv")
```

```python
print("Fill null values with empty string...")
train['question1'].fillna('', inplace=True)
train['question2'].fillna('', inplace=True)
```

## Word Cloud

Do the two questions share similar words?

```python
qid_dict = {}

for i,series in train.iterrows():
    if series['qid1'] not in qid_dict:
        qid_dict[series['qid1']] = series['question1']
    if series['qid2'] not in qid_dict:
        qid_dict[series['qid2']] = series['question2']
```

```python
txt = []

for key, value in qid_dict.items():
    txt.append(str(value))
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()
```

```python
import re

def extract_word(df, colname='question1'):
    q_word = ''
    for idx, row in df.iterrows():
        for word in row[colname].split():
            q_word += (re.sub('[^a-zA-Z]+','',word.lower())) + ' '
    return q_word
```

```python
q1_word = extract_word(train, colname='question1')
q2_word = extract_word(train, colname='question2')
```

```python
show_wordcloud(q1_word, title='Question 1')
```

```python
show_wordcloud(q2_word, title='Question 2')
```

```python
import nltk

words = [word.lower() for word in q1_word.split(' ') if word.lower() not in stopwords and word.isalpha()]
freqdist = nltk.FreqDist(words)
plt.figure(figsize=(16,5))
plt.title("Question 1")
freqdist.plot(30)
```

```python
words = [word.lower() for word in q2_word.split(' ') if word.lower() not in stopwords and word.isalpha()]
freqdist = nltk.FreqDist(words)
plt.figure(figsize=(16,5))
plt.title("Question 2")
freqdist.plot(30)
```

# Data Exploration

We'll check if there is any interesting features shared on the two classes.

## Common words

```python
from scipy.stats import norm, skewnorm

f, ax = plt.subplots(figsize=(8,6))

sns.distplot(train[(train['is_duplicate'] == 1)]['num_common_words'],
             kde=False,
             fit=norm,
             fit_kws={'linestyle':'--', 'color':'green','label':'Duplicate'},
             label='Duplicate'
            )
sns.distplot(train[(train['is_duplicate'] == 0)]['num_common_words'],
            kde=False,
            fit=norm,
            fit_kws={'linestyle':'--', 'color':'red','label':'Not duplicate'},
            label='Not duplicate'
            )

plt.xlabel("Number of common words")
plt.ylabel("Density")
plt.legend(loc='upper right')
ax.set_xlim(0,)
plt.show();
```

```python
print("Avg number of common words on duplicate pairs: {}".format(train[(train['is_duplicate'] == 1)]['num_common_words'].mean()))
print("Avg number of common words on non-duplicate pairs: {}".format(train[(train['is_duplicate'] == 0)]['num_common_words'].mean()))
```

```python
from scipy.stats import norm, skewnorm

f, ax = plt.subplots(figsize=(8,6))

sns.distplot(train[(train['is_duplicate'] == 1)]['pct_common_words'],
             kde=False,
             fit=norm,
             fit_kws={'linestyle':'--', 'color':'red','label':'Duplicate'},
             hist_kws={'color':'red'},
             label='Duplicate'
            )
sns.distplot(train[(train['is_duplicate'] == 0)]['pct_common_words'],
            kde=False,
            fit=skewnorm,
            fit_kws={'linestyle':'--', 'color':'green','label':'Not duplicate'},
            hist_kws={'color':'green'},
            label='Not duplicate'
            )

plt.xlabel("Percentage of common words")
plt.ylabel("Density")
plt.legend(loc='upper right')
ax.set_xlim(0,)
plt.show();
```

```python
print("Avg % of common words on duplicate pairs: {}".format(train[(train['is_duplicate'] == 1)]['pct_common_words'].mean()))
print("Avg % of common words on non-duplicate pairs: {}".format(train[(train['is_duplicate'] == 0)]['pct_common_words'].mean()))
```

## Jaccard distance

```python
f, ax = plt.subplots(figsize=(8,6))

sns.distplot(
    train[(train['is_duplicate'] == 1)]['jaccard_dist'],
    kde_kws={'linestyle':'--', 'color':'red','label':'Duplicate'},
    hist_kws={'color':'red'},
    label='Duplicate'
)

sns.distplot(
    train[(train['is_duplicate'] == 0)]['jaccard_dist'],
    kde_kws={'linestyle':'--', 'color':'green','label':'Not duplicate'},
    hist_kws={'color':'green'},
    label='Not duplicate'
)

plt.xlabel("Jaccard distance")
plt.ylabel("Density")
plt.legend(loc='upper right')
# plt.title("Does duplicate question have different Jaccard distance distribution?")
plt.show()

print("Avg Jaccard distance on duplicate pairs: {}".format(train[(train['is_duplicate'] == 1)]['jaccard_dist'].mean()))
print("Avg Jaccard distance on non-duplicate pairs: {}".format(train[(train['is_duplicate'] == 0)]['jaccard_dist'].mean()))
```

It seems like most records with Jaccard distance > 0.2 are duplicate pairs.

## Cosine distance

```python
f, ax = plt.subplots(figsize=(8,6))

sns.distplot(
    train[(train['is_duplicate'] == 1)]['cosine_dist'],
    kde_kws={'color':'red'},
    hist_kws={'color':'red'},
    label='Duplicate'
)

sns.distplot(
    train[(train['is_duplicate'] == 0)]['cosine_dist'],
    kde_kws={'color':'green'},
    hist_kws={'color':'green'},
    label='Not duplicate'
)

plt.xlabel("Cosine distance")
plt.ylabel("Density")
plt.legend(loc='upper right')
plt.title("Does duplicate question have different cosine distance distribution?")
plt.show()
```

## Edit distance first & last word

```python
f, ax = plt.subplots(figsize=(8,6), nrows=2)
plt.subplots_adjust(hspace=.5)

sns.boxplot(
    data=train[(train['question2'] != '')],
    x='edit_dist_first_word',
    y='is_duplicate',
    orient='h',
    palette='tab20',
    ax=ax[0]
)

sns.boxplot(
    data=train[(train['question2'] != '')],
    x='edit_dist_last_word',
    y='is_duplicate',
    orient='h',
    palette='tab20',
    ax=ax[1]
)

plt.show()
```

```python
train.groupby('is_duplicate').describe()[['edit_dist_first_word','edit_dist_last_word']]
```

Duplicate pairs tend to have more similar first and last word than non-duplicate pairs.
